```python
def DFS(graph, root):
    visited = []
    stack = [root]
    # DFS는 이전에 방문했었던, 노드들을 거쳐야 하므로 stack으로 구현

    while stack:
        print('now stack : ', stack)
        n = stack.pop()# list우측 제거(stack_pop)
        print('visited ', visited)
        print('check ? ', n)
        if n not in visited:
            visited.append(n)
            if n in graph:
                print('간선 판단 [n으로부터 뻗어있는]-[방문한] : {} - {}'.format(set(graph[n]), set(visited)))
                temp = list(set(graph[n]) - set(visited))
                temp.sort(reverse=True)
                # stack에 들어갈때는, reverse시켜서
                # 가장 먼저 들어간 수를 마지막에 탐색하게끔..
                    # [5,4,3,2,1] 순서로 들어감 -> 탐색 순서 : 1,2,3,4,5
                print('방문하지 않은 간선: ',temp)
                stack+=temp
        print('\tnext\n')
    return " ".join(str(i) for i in visited)

def BFS(graph, root):
    visited = []
    queue = deque([root])

    while queue:
        n = queue.popleft()
        # deque의 왼쪽을 계속 제거, 그 노드가 만약 방문하지 않은 경우 if문 수행, 
        # 아니라면 계속 왼쪽 제거
        if n not in visited:
            visited.append(n)
            if n in graph:
                temp = list(set(graph[n]) - set(visited))
                # 방문하지 않은 간선(집합 연산)을 list로 temp에 저장
                temp.sort() # 정렬(bfs)
                print(temp)
                queue += temp
        else: continue
    return " ".join(str(i) for i in visited)

print(DFS(graph, start))
print(BFS(graph, start))
```
